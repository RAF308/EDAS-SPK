```python
# Required Libraries
import numpy as np

from pyDecision.algorithm import edas_method
```

```python
# EDAS

# Weights
weights = [0.3, 0.2, 0.2, 0.15, 0.15]

# Load Criterion Type: 'max' or 'min'
criterion_type = ['max', 'max', 'max', 'min', 'min']

# Dataset
dataset = np.array([
                [0.5, 1, 0.7, 0.7, 0.8], #a1
                [0.8, 0.7, 1, 0.5, 1], #a2
                [1, 0.3, 0.4, 0.7, 1], #a3
                [0.2, 1, 0.5, 0.9, 0.7], #a4
                [1, 0.7, 0.4, 0.7, 1] #a5
                ])
```

```python
# Call EDAS Function
rank = edas_method(dataset, criterion_type, weights, graph = True, verbose = True)
```
